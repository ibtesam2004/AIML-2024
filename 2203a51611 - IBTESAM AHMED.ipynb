```python
import numpy as np
# NumPy array
my_array = np.array([1, 2, 3])
```

```python
# Sine function
result_sin = np.sin(my_array)
print("NumPy sin():", result_sin)
```

```python
# Natural logarithm
result_log = np.log(my_array)
print("\nNumPy log():", result_log)
```

```python
# Base-2 logarithm
result_log2 = np.log2(my_array)
print("\nNumPy log2():", result_log2)
```

```python
# Exponential function (e^x)
result_exp = np.exp(my_array)
print("\nNumPy exp():", result_exp)
```

```python
import matplotlib.pyplot as plt

# 1. Generate Data set for Health Care
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
```

```python
# 2. Scatter Plot
plt.figure(figsize=(5,5))
plt.scatter(age, height, s=weight, c='blue', alpha=0.7, label='Weight')
plt.title('Scatter Plot of Age vs Height with Weight')
plt.xlabel('Age')
plt.ylabel('Height')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. Bar Chart
plt.figure(figsize=(8, 6))
bar_width = 0.3
index = range(len(people))
plt.bar(index, weight, bar_width, color='green', label='Weight')
plt.title('Bar Chart of Weight for Each Person')
plt.xlabel('People')
plt.ylabel('Weight')
plt.xticks(index, people)
plt.legend()
plt.show()
```

```python
# 4. Histogram
plt.figure(figsize=(8, 6))
plt.hist(age, bins=5, color='orange', edgecolor='black')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
